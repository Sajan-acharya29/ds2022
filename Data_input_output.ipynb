```python
df = pd.read_csv("test_csv_file")
df
df.info()               #to get the more infomration about your data from columns

new_columns = (['NameBank', 'City', 'State', 'Cert',
       'Acquiring InstitutionAI', 'DateClosing', 'Fund'])
df.columns = new_columns
df.columns
df.head(2)
```

```python
df['DateClosing'].dtypes
```

```python
df['New_DateClosing'] = pd.to_datetime(df['DateClosing'])
df.head(2)
```

```python
df.drop('DateClosing', axis=1, inplace = True)
```

```python
df.head(2)
```

```python
df['New_DateClosing'].dtypes
```

```python
#converting the datetime object to a string

df['New_DateClosing'] = df['New_DateClosing'].astype('string')
```

```python
df['New_DateClosing'].dtypes
```

```python
df['year']= df['New_DateClosing'].apply(lambda x: x.split('-')[0])
df['month']= df['New_DateClosing'].apply(lambda x: x.split('-')[1])
df['date']= df['New_DateClosing'].apply(lambda x: x.split('-')[2])
```
